```python
import pandas as pd
import matplotlib.pyplot as plt
import json
import pprint
import datetime as dt
import numpy as np
```

```python
url = 'https://query.data.world/s/o6que7zbpjnykcz3ayncuhwkgsb3dp'
df = pd.DataFrame(pd.read_csv(url))
df = df.loc[df.COUNTRY_ALPHA_3_CODE == 'USA']
# df = df.loc[df['REPORT_DATE'] != '2020-07-30']
df.head()
```

```python
df['county_state'] = df['COUNTY_NAME'] + ', ' + df['PROVINCE_STATE_NAME']

df = df[['county_state',
        'REPORT_DATE',
        'PEOPLE_POSITIVE_NEW_CASES_COUNT',
        'PEOPLE_POSITIVE_CASES_COUNT',
        'PEOPLE_DEATH_NEW_COUNT',
        'PEOPLE_DEATH_COUNT']]

df = df.rename(columns={'REPORT_DATE':'date',
                        'PEOPLE_POSITIVE_NEW_CASES_COUNT':'new_cases',
                        'PEOPLE_POSITIVE_CASES_COUNT':'total_cases',
                        'PEOPLE_DEATH_NEW_COUNT':'new_deaths',
                        'PEOPLE_DEATH_COUNT':'total_deaths'})

df
```

```python
county_list = df['county_state'].unique().tolist()
county_list.insert(0,'USA - Mean')
county_list.insert(0,'USA - Sum')
county_list
```

```python
def county_select(county_name):
    
    '''
    A function that takes a county name and returns COVID data in JSON form
    '''
    
    if county_name == 'USA - Sum':
        total_df = df.groupby('date')[['new_cases','total_cases','new_deaths','total_deaths']].sum()
        output_json = {'date':total_df.index.to_list(),
                       'new_cases':total_df.new_cases.to_list(),
                       'total_cases':total_df.total_cases.to_list(),
                       'new_deaths':total_df.new_deaths.to_list(),
                       'total_deaths':total_df.total_deaths.to_list()
                      }
        
    elif county_name == 'USA - Mean':
        avg_df = df.groupby('date')[['new_cases','total_cases','new_deaths','total_deaths']].mean()
        output_json = {'date':avg_df.index.to_list(),
                       'new_cases':avg_df.new_cases.to_list(),
                       'total_cases':avg_df.total_cases.to_list(),
                       'new_deaths':avg_df.new_deaths.to_list(),
                       'total_deaths':avg_df.total_deaths.to_list()
                      }
        
    else:
        county_df = df.loc[df.county_state==county_name]
        output_json = {'date':county_df.date.to_list(),
                       'new_cases':county_df.new_cases.to_list(),
                       'total_cases':county_df.total_cases.to_list(),
                       'new_deaths':county_df.new_deaths.to_list(),
                       'total_deaths':county_df.total_deaths.to_list()
                      }
        
    return output_json

county_select('USA - Mean')
```

```python
# Limit data to Travis County
travis_co = df.loc[(df['PROVINCE_STATE_NAME']=='Texas') & (df['COUNTY_NAME']=='Travis')].reset_index()

travis_co = travis_co[['REPORT_DATE',
                       'PEOPLE_POSITIVE_NEW_CASES_COUNT',
                       'PEOPLE_POSITIVE_CASES_COUNT',
                       'PEOPLE_DEATH_NEW_COUNT',
                       'PEOPLE_DEATH_COUNT']]

travis_co = travis_co.rename(columns={'REPORT_DATE':'date',
                              'PEOPLE_POSITIVE_NEW_CASES_COUNT':'new_cases',
                              'PEOPLE_POSITIVE_CASES_COUNT':'total_cases',
                              'PEOPLE_DEATH_NEW_COUNT':'new_deaths',
                              'PEOPLE_DEATH_COUNT':'total_deaths'})

travis_co
```
